Getting data for last 2 month

```python
def convert_date(date_str):
    try:
        return pd.to_datetime(date_str, format='%Y-%m-%d %H:%M:%S')
    except ValueError:
        return pd.NaT

df['date'] = df['date'].apply(convert_date)  # convert the date column to datetime
df = df.dropna(subset=['date'])  # drop rows with NaT in the date column

last_date_in_data = df['date'].iloc[-1]  # get the date in the last row of the dataset
one_year_before_last_date = last_date_in_data - pd.DateOffset(month=2)  # get the date one year before the last date
df_last_year = df[df['date'] >= one_year_before_last_date]  # filter the data for the last 2 month
```

Create a merged dataset of the Bitcoin price at the time the tweet was posted and the price 12 hours later.

```python
from datetime import timedelta

def add_start_end_price(df1, btc_df):
    # Ensure the 'date' columns are in datetime format
    df1['date'] = pd.to_datetime(df1['date'])
    btc_df['date'] = pd.to_datetime(btc_df['date'])

    # Sort btc_df by date
    btc_df = btc_df.sort_values('date')

    # Initialize 'start' and 'end' columns
    df1['start'] = 0
    df1['end'] = 0

    for i, row in df1.iterrows():
        # Find the closest date in btc_df
        nearest_date = btc_df.iloc[(btc_df['date'] - row['date']).abs().argsort()[:1]]
        
        # Add the price at the nearest date to the 'start' column
        df1.at[i, 'start'] = nearest_date['price'].values[0]

        # Add 12 hours to the nearest date
        date_12h_later = pd.Timestamp(nearest_date['date'].values[0]) + timedelta(hours=12)

        # Find the price 12 hours later
        price_12h_later = btc_df[btc_df['date'] == date_12h_later]['price']

        # If the price 12 hours later exists, add it to the 'end' column
        if not price_12h_later.empty:
            df1.at[i, 'end'] = price_12h_later.values[0]

    return df1
```

Downloading models for sentiment analysis

```python
from transformers import AutoModelForSequenceClassification, pipeline
from tqdm import tqdm

# Define the models
models = [
    'ProsusAI/finbert',
    'mrm8488/distilroberta-finetuned-financial-news-sentiment-analysis',
    'lxyuan/distilbert-base-multilingual-cased-sentiments-student',
]

# Load the models and get label mappings
pipelines = [pipeline('text-classification', model=model_name) for model_name in models]
label_mappings = [{v: k for k, v in AutoModelForSequenceClassification.from_pretrained(model_name).config.id2label.items()} for model_name in models]

label_mappings
```

```python
for i, row in tqdm(df.iterrows(), total=df.shape[0]):
    text = row['text']
    
#     # Get sentiment scores and add to DataFrame
    for j, pipe in enumerate(pipelines):
        scores = pipe(text)[0]
        for label, score in scores.items():
            df.at[i, f'score{j+1}_{label}'] = score
```

```python
# Replace the sentiment labels with numbers
for j, label_mapping in enumerate(label_mappings):
        df[f'score{j+1}_label'] = df[f'score{j+1}_label'].map(label_mappings[j])
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import make_pipeline
from xgboost import XGBClassifier


# Calculate the sign of the difference between 'end' and 'start' columns and add 1 then //2
df['end_start_diff_sign'] = (np.sign(df['end'] - df['start']) + 1) // 2

# Define features and target variable
X = df[['start', 'score1_label', 'score1_score', 'score2_label', 'score2_score', 'score3_label', 'score3_score']]
y = df['end_start_diff_sign']

# Split the data into training and test sets (first 80% for training, last 20% for testing)
train_size = int(0.8 * len(df))
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]


pipeline = make_pipeline(
    XGBClassifier()
)

# Train the model
pipeline.fit(X_train, y_train)

# Test the model
accuracy = pipeline.score(X_test, y_test)
print(f'Test accuracy: {accuracy}')
```

```python
# Initialize a variable to hold  balance
balance = 100

# Iterate over the DataFrame with a step of 5
for i in range(0, len(df), 5):
    # Check if there are less than 5 rows left
    if i + 5 > len(df):
        break

    # Get the current group of 5 rows
    group = checkingdf.iloc[i:i+5]

    # Count the number of predictions that say the price will go up
    num_up_predictions = group[group['predictions'] == 1].count()

    # If at least 3 out of 5 predictions say the price will go up
    if num_up_predictions['predictions'] >= 3:
        # Calculate the new balance
        balance = balance * (1 + (group.iloc[0]['end'] - group.iloc[0]['start']) / group.iloc[0]['start'])

        # Skip the next 12*5 rows
        i += 12 * 5

print(f'Final balance: {balance}')
```
